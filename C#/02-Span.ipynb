# **Span**
-   Span<T> is a ref struct that is allocated on the stack instead of heap.

### Span memory Allocation


```python
var spandex = new Spandex();

var (month, day, year) = spandex.ParseDateSpan();

Console.WriteLine($"The date is {month}/{day}/{year}");

public class Spandex
{
    private readonly string _dateAsText = "05 11 2023";

    public (int month, int day, int year) ParseDateString()
    {
        var monthString = _dateAsText.Substring(0, 2);

        var dayString = _dateAsText.Substring(3, 2);

        var yearString = _dateAsText.Substring(6);

        var month = int.Parse(monthString);

        var day = int.Parse(dayString);

        var year = int.Parse(yearString);

        return (month, day, year);
    }

    public (int month, int day, int year) ParseDateSpan()
    {
        ReadOnlySpan<char> dateAsSpan = _dateAsText;

        var monthSpan = dateAsSpan.Slice(0, 2);

        var daySpan = dateAsSpan.Slice(3, 2);

        var yearSpan = dateAsSpan.Slice(6, 4);

        var month = int.Parse(monthSpan);

        var day = int.Parse(daySpan);

        var year = int.Parse(yearSpan);

        return (month, day, year); 
    }
}
```
